## import Libraries

```python
import pandas as pd
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
```

## Load the Data

```python
data = pd.read_csv('globalterrorismdb.csv',encoding='latin1')
data.head()
```

```python
data.describe()
```

```python
data.shape
```

```python
data.columns.values
```

```python
Data =data[['iyear', 'imonth', 'iday','country', 'country_txt', 'region',
            'region_txt','provstate', 'city','targtype1','targtype1_txt',
            'attacktype1', 'attacktype1_txt','nkill','nwound', 'gname' ,'weaptype1_txt','weaptype1']]
Data.head()
```

## Exploratory Data Analysis

```python
plt.figure(figsize=(20,10))
sns.heatmap(Data.corr(),annot= True)
```

```python
plt.figure(figsize=(20,10))
plt.xticks(size=10)
sns.countplot(Data['iyear'])
print(Data['iyear'].value_counts())
```

```python
plt.figure(figsize=(50,25))
sns.countplot(Data['country_txt'])
print (Data['country_txt'].value_counts())
```

```python
plt.figure(figsize=(30,10))
sns.countplot(Data['region_txt'])
print (Data['region_txt'].value_counts())
```

```python
plt.figure(figsize=(50,20))
plt.xticks(size=25)
sns.countplot(Data['targtype1_txt'])
print(Data['targtype1_txt'].value_counts())
```

```python
plt.figure(figsize=(50,20))
plt.xticks(size=25)
sns.countplot(Data['attacktype1_txt'])
print(Data['attacktype1_txt'].value_counts())
```

```python
plt.figure(figsize=(50,20))
plt.xticks(size=25)
sns.countplot(Data['weaptype1_txt'])
print(Data['weaptype1_txt'].value_counts())
```
