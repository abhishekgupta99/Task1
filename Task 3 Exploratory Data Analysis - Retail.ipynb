## Exploratory Data Analysis

```python
Data.corr()
sns.heatmap(Data.corr(),annot=True)
```

```python
sns.pairplot(Data)
```

```python
Total_Profit = Data['Profit'].sum()
Total_Profit 
```

```python
Total_Sales = Data['Sales'].sum()
Total_Sales 
```

```python
plt.figure(figsize=(20,10))
sns.countplot(Data['Segment'])
print(Data['Segment'].value_counts())
```

```python
plt.figure(figsize=(20,10))
sns.countplot(Data['Category'])
print(Data['Category'].value_counts())
```

```python
plt.figure(figsize=(20,10))
sns.countplot(Data['Sub-Category'])
print(Data['Sub-Category'].value_counts())
```

```python
plt.figure(figsize=(20,10))
sns.countplot(Data['Region'],hue=('Category'),data=Data)
```

```python
plt.figure(figsize=(20,10))
sns.countplot(Data['Sub-Category'],hue=('Region'),data=Data)
```

```python
Data1 = Data.groupby(['Region']).sum()
Data1
```

```python
Data1=Data1.sort_values(['Sales'],ascending=False)
sns.barplot(x='Sales',y=Data1.index,data=Data1)
plt.show()
```

```python
Data1=Data1.sort_values(['Profit'],ascending=False)
sns.barplot(x='Profit',y=Data1.index,data=Data1)
plt.show()
```

```python
Data2 = Data.groupby(['State']).sum()
Data2
```

```python
plt.figure(figsize=(20,50))
Data2=Data2.sort_values(['Sales'],ascending=False)
sns.barplot(x='Sales',y=Data2.index,data=Data2)
plt.show()
```

```python
plt.figure(figsize=(20,15))
plt.xticks(size=15)
plt.yticks(size=15)
Data2=Data2.sort_values(['Profit'],ascending=False)
sns.barplot(x='Profit',y=Data2.index,data=Data2)
plt.show()
```
